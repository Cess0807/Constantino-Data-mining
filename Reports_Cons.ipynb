<h1> 15: Correlation Between Various Expenditures

```python
# Select relevant expenditure columns for correlation
expenditure_cols = [col for col in expenditure.columns if 'Expenditure' in col]
corr_matrix = expenditure[expenditure_cols].corr()

# Generate heatmap for correlation between expenditures
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Various Household Expenditures")
plt.tight_layout()
plt.show()
```

<h1> Insight 15 <h1>

<h2> The correlation heatmap provides a comprehensive view of relationships between spending categories. Total expenditure shows strong positive correlations with food (0.91), housing (0.75), and transportation (0.76), confirming these as core household budget components. Interestingly, restaurant and alcoholic beverage expenditures show stronger correlations with each other (0.70) than with total food spending, suggesting discretionary social spending patterns. Medical care displays relatively weak correlations with most categories (mostly below 0.40) </h2>

<h1> 20: Appliance Ownership Distribution
